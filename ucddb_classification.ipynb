# UCD sleep dataset: Evaluation of different classification strategies on a subset of the data

## Original dataset and previous work

In Langkvist (2012), "Sleep Stage Classifiction Using Unsupervised Feature Learning", the authors wanted to classify different sleep stages using physiological information only. The raw data is freely available [online](http://physionet.org/pn3/ucddb/):
* 25 participants
* ~7 hours of recording per participant
* Physiological signals
 * EEG (C3, C4)
 * EMG (face)
 * EOG (2 channels)
* 128 Hz (EEG), 64 Hz (EOG, EMG)
* 5 sleep stages

The authors tried three different architectures: 
1. Feature extraction, Feature selection, PCA, GMM and HMM
2. Feature extraction, DBN and HMM 
3. DBN and HMM

Findings suggest that while all three methods yielded comparable results (64-72% on average in the 5-class classification task), the second architecture was slightly better.


## Task and proposed approach

This notebook will explore different classification strategies on a subset of the dataset used in Langkvist (2012) (data and MATLAB code available [here](http://aass.oru.se/~mlt/)). Specifically, this subset is composed of:
* 2 classes: drowsy (0) and awake (1)
* 42,570 observations (computed on non-overlapping 1-second windows)
* 7 types of features
 * EEG frequency band powers: EEG_delta, EEG_theta, EEG_alpha, EEG_beta, EEG_gamma
 * Standard deviation of an EOG channel: EOG_std
 * EEG fractal dimension exponent: EEG_fractal_exponent

Drowsy and awake labels are obtained from the original dataset (first stage of sleep and awake conditions), from which five subjects are extracted.

Unlike Langkvist (2012), this subset does not contain temporal continuity. Indeed, the sequence of observations is not structured like it previously was. Therefore, the use of a Hidden Markov Model is not appropriate in this case.

The goal of this project is thus to obtain the highest possible accuracy on this 2-class dataset. Additionnally, it should be possible to interpret the trained classifiers to understand which features are relevant and how.

The proposed approach follows the standard workflow of a classification task:
1. Exploration of the dataset
2. (If necessary) Preprocessing of the dataset
3. Model selection between many different classifiers: 
 1. (If necessary) Hyperparameter optimization on training set with cross-validation
 2. Evaluation of the optimal classifier on a held-out test set
 
To increase the performance, different strategies can be used (and will be implemented if time allows):
* dataset augmentation
* adding more data from the original dataset
 * use the whole 25 subjects
 * use the whole 28 features
* relying on unsupervised pretraining for extracting features (from raw data or precomputed features)
* using deep architectures
* using convolutional neural networks on EEG spectrogram

These steps will be followed using [pandas](http://pandas.pydata.org/) and [scikit-learn](http://scikit-learn.org/stable/) as primary tools.

# 1. Exploration of the dataset

Let's load the data, put it in a pandas DataFrame and print a summary:

```python
import scipy.io
import numpy as np
import pandas as pd
```

```python
filepath = 'C:\\data\\ucddb\\drowsy_awake_7feat.mat'
data = scipy.io.loadmat(filepath)
    
# 'data' should contain fields 'CLASS', 'FEAT' and 'feat_names'
feat = data['FEAT']
labels = data['CLASS']
feat_names = data['feat_names']

# List comprehension for cleaning the chList
featNames_clean = [str(''.join(letter)) for letter_array in feat_names[0] for letter in letter_array]

assert (feat.shape[1] == feat_names.shape[1]), 'The number of features is not unanimous between feats and featNames.'    

# Make a pandas dataframe
feats = pd.DataFrame(data = feat, columns = featNames_clean)
feats['label'] = pd.Series(data = np.reshape(labels,(-1,)))

feats.describe()
```

The features were most likely z-score normalized on the complete dataset, which is why the means and standard deviations are close to 0 and 1, respectively, but still slightly off. Additional normalization is thus not necessary.

To get a sense of the features' distributions, we can start by plotting the corresponding boxplots:

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
feats.groupby('label').boxplot(vert=False)

for i in range(7):
    ax = plt.figure()
    featName = feats.columns[i]
    feats.groupby('label')[featName].hist(alpha=0.4, bins=100)
    plt.xlabel(featName)
    plt.ylabel('Number of observations')
    plt.legend(['drowsy','awake'])
```

Apart from EOG_std which is highly skewed to the right, the features look more or less normally distributed. 
We can observe which features are the most likely to be relevant by computing the Pearson's correlation between them and the labels:

```python
# Correlation with the labels
plt.figure()
feats.corr()['label'].plot(kind='barh')
plt.title('Pearson correlation of the features with the labels')
plt.xlim(xmax = 1.1, xmin = -0.5);
```

EEG_fractal_exponent is the most highly correlated, while the powers in alpha and theta bands are very poorly correlated. Eventually, these two features might be dropped to simplify even further the classification task.

Let's take a quick look at what these features look like in 2D:

```python
# Scatter plot of features
colors = np.where(feats.label > 0.5, 'r', 'b')
feats.plot(kind='scatter', x='EEG_fractal_exponent', y='EEG_gamma', alpha=0.3, c=colors, s=50)
plt.legend(['drowsy','awake'])
feats.plot(kind='scatter', x='EOG_std', y='EEG_beta', alpha=0.3, c=colors, s=50)
plt.legend(['drowsy','awake'])
```

Finally, to get an idea of how variance is explained in the dataset and since it was used as a preprocessing step by the original authors, we can decompose the features using a PCA, and plot the resulting components:

```python
from sklearn import decomposition
```

```python
# Plot principal components
pca = decomposition.PCA(n_components=3)
pca.fit(feats.iloc[:,:-1])
PCs = pca.transform(feats.iloc[:,:-1])
PCs = pd.DataFrame(PCs, columns=['PC1','PC2','PC3'])
PCs['label'] = feats['label']
PCs.plot(kind='scatter', x='PC1', y='PC2', alpha=0.3, label='Drowsy', c=colors, s=50)
PCs.plot(kind='scatter', x='PC3', y='PC2', alpha=0.3, label='Drowsy', c=colors, s=50)

# Explained variance
print('Explained variance by 3 first components: %0.3f'%np.sum(pca.explained_variance_ratio_))
```

The first component explains the most variance, but the second and third components seem more informative, as the difference between classes is more apparent. This is to be taken into account if PCA is used in the classification pipeline.

## 2. Model evaluation

Since we're going to compare multiple classifiers on the same data, it is important to get an unbiased estimate of their individual accuracy. For this reason, we will split the dataset in three parts:
1. Training set
2. Validation set
3. Testing set

To minimize the loss of training data, the first two sets will be separated using k-fold cross-validation. This will allow us to train a classifier with a set of hyperparameters on k-1 folds, and to validate it on the remaining fold (if need be). The testing set, however, will be kept unseen during the training so all classifiers can be compared. Eventually, a nested two-stage cross-validation could be done to avoid completely loosing the testing set data.

Let's keep 15% of the data for testing, and divide the remaining 85% of data in 10 stratified folds (meaning that a similar percentage of observations from each class is kept in each fold). 

```python
from sklearn import cross_validation
```

```python
rng = 42 # Set random number

# Separate data in train and test sets
X_train, X_test, y_train, y_test = cross_validation.train_test_split(
                                    feats.iloc[:,:-1], feats.label, test_size=0.15, random_state=rng)

# Separate train set in stratified folds for cross-validation
cv = cross_validation.StratifiedKFold(y_train, n_folds=10)

# Create a DataFrame to keep the results
clfs = np.zeros((2,9)) # row 1: validation acc. mean, row 2: test acc.
```

Before training our models, let's make a helper function that will take care of training and testing, and of displaying the results:

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.grid_search import GridSearchCV
```

```python
def gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test):
    """First perform a grid search of hyperparameters contained in param_grid, then test
    on the test set. Print out some verbose and the results.
    
    Output the best trained model.
    
    Adapted from http://scikit-learn.org/stable/auto_examples/grid_search_digits.html#example-grid-search-digits-py
    """
    
    score = 'accuracy'
    
    print('# Tuning hyper-parameters for %s' %score)
    print('')

    clf = GridSearchCV(clf, param_grid, cv=cv, scoring=score, n_jobs=-1, verbose=True)
    clf.fit(X_train, y_train)

    print('Best parameters set found on development set:')
    print('')
    print(clf.best_estimator_)
    print('')
    print('Grid scores on development set:')
    print('')
    for params, mean_score, scores in clf.grid_scores_:
        print('%0.3f (+/-%0.03f) for %r'
              % (mean_score, scores.std() / 2, params))
    print('')

    print('Detailed classification report:')
    print('')
    print('The model is trained on the development set.')
    print('The scores are computed on the full test set.')
    print('')
    
    y_pred = clf.predict(X_test)
    print(classification_report(y_test, y_pred))
    print('Accuracy: %0.3f'%accuracy_score(y_test, y_pred))
    print('')
    
    # Get best model scores
    bestclf = clf.best_estimator_
    y_pred = bestclf.predict(X_train)
    train_acc = accuracy_score(y_train, y_pred)

    y_pred = bestclf.predict(X_test)
    test_acc = accuracy_score(y_test, y_pred)

    return bestclf, train_acc, test_acc
```

### a. Random classification

This one is obvious, but let's start with this to see what a random classifier would give us on this dataset.

```python
from sklearn.dummy import DummyClassifier
```

```python
print('Classifier 0: Random')
print('')

# see http://scikit-learn.org/stable/auto_examples/plot_permutation_test_for_classification.html#example-plot-permutation-test-for-classification-py
randclf = DummyClassifier(strategy='uniform',random_state=rng)
randclf.fit(X_train, y_train)

y_pred = randclf.predict(X_train)
train_acc = accuracy_score(y_train, y_pred)

y_pred = randclf.predict(X_test)
test_acc = accuracy_score(y_test, y_pred)

print(classification_report(y_test, y_pred))
print('Accuracy: %0.3f'%test_acc)
print('')

clfs[0,0] = train_acc
clfs[1,0] = test_acc
```

Unsurprisingly, this one gives us a 50% chance of guessing the right class.

### b. LDA

Linear Discriminant Analysis is a mainstream linear classifier, especially in the world of BCIs. Knowing our features are not too far from being normally distributed, LDA should give us a good performance.
No hyperparameters need to be optimized for this classifier, so we can use the whole training set (without cross-validation) to train it.

```python
from sklearn import lda
```

```python
print('Classifier 1: LDA')
print('')

LDAclf = lda.LDA()
LDAclf.fit(X_train, y_train)

y_pred = LDAclf.predict(X_train)
train_acc = accuracy_score(y_train, y_pred)

y_pred = LDAclf.predict(X_test)
test_acc = accuracy_score(y_test, y_pred)

print(classification_report(y_test, y_pred)) 
print('Accuracy: %0.3f'%accuracy_score(y_test, y_pred))   
print('')

clfs[0,1] = train_acc
clfs[1,1] = test_acc
```

Our first benchmark is 71.3%.

### c. Logistic regression

As a second widely used linear classifier, logistic regression is more robust to outliers than LDA (which doesn't seem like a huge problem in our dataset anyway). This time, we will optimize the hyperparameter C (controlling the strength of the regularization; lower is stronger) and the type of regularization (either l1 or l2).

```python
from sklearn import linear_model
```

```python
print('Classifier 2: Logistic regression')
print('')    

clf = linear_model.LogisticRegression(random_state=rng)

param_grid = [{'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['l2', 'l1']},]

bestLRclf, train_acc, test_acc = gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test)

print('')

clfs[0,2] = train_acc
clfs[1,2] = test_acc
```

Almost all models gave very similar validation accuracies (~72.7%). The best model found was the one with C=0.1 and l2-penalty, and scored 72.8% on the test set.
Let's take a look at the coefficients for each feature:

```python
LRcoef = pd.DataFrame(bestLRclf.coef_, columns=feats.columns[:-1])
LRcoef.T.plot(kind='barh')
plt.ylabel('LR coefficient value')
```

We expect EEG_theta and EEG_alpha coefficient values to be low because they were previously shown to be poorly correlated with the labels. Surprisingly, the coefficients learned by the logistic regression of EEG_theta and EEG_alpha rather were higher than EEG_beta and EEG_gamma, meaning they were found to be more informative by the LR. 

Moreover, the correlation analysis previously showed that EEG_gamma, EEG_beta and EEG_alpha are positively correlated with the labels; however, the LR found a negative weights for these features, which is counter-intuitive. Further analysis of the model might be necessary to explain this effect.

### d. Linear and kernel SVMs

The last linear classifier we'll use is the Support Vector Machine. The kernel versions of the SVM are also very powerful, so we'll run the cross-validation over two hyperparameters:
* C: soft margin coefficient (the higher, the more penalty is given to errors)
* gamma: (for the RBF kernel) controls the size of the kernel

The following code can take a long time to run (>60 min on a mobile i7 CPU with 2 cores) because of the cross-validation and the large amount of data (even with the tolerance parameter of the SVM training function set to 0.1).

Let's start with a linear SVM:

```python
from sklearn import svm
```

```python
print('Classifier 3: Linear SVM')
print('')

clf = svm.LinearSVC(random_state=rng)

# Set up the hyperparameter search    
param_grid = [{'C': [0.1, 1, 5, 10, 25, 50, 100]}]    

bestLinearSVMclf, train_acc, test_acc = gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test)

print('')

clfs[0,3] = train_acc
clfs[1,3] = test_acc
```

A similar performance as the best Linear Regression model is obtained for the linear SVM with C=10, at 72.8% validation accuracy. Let's take a look at the weights of this best model:

```python
LinearSVMcoef = pd.DataFrame(bestLinearSVMclf.coef_, columns=feats.columns[:-1])
LinearSVMcoef.T.plot(kind='barh')
plt.ylabel('Linear SVM coefficient value')
```

The relative weight for each feature is very similar to what was obtained with Logistic Regression as well.

Let's now try with a kernel SVM:

```python
print('Classifier 4: Kernel SVM')
print('')

clf = svm.SVC(random_state=rng, tol=0.1)

# Set up the hyperparameter search    
param_grid = [{'C': [0.1, 1, 10, 100, 1000], 'gamma': [0.01, 0.001, 0.0001], 'kernel': ['rbf']}]    

bestKernelSVMclf, train_acc, test_acc = gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test)

print('')

clfs[0,4] = train_acc
clfs[1,4] = test_acc
```

The model with C=1000 and gamma=0.01 gave the best test accuracy at 75.6%. Even better results could probably be obtained by using larger C and gamma values.

### e. K-nearest neighbors

The k-nearest neighbors algorithm is non-parametric, and can take a lot of memory since it works by remembering all the points in the dataset. However, since our dataset is low-dimensional (7 features), this is not a huge issue; additionnally, this means the curse-of-dimensionality should not be a problem.

Let's try different values of k (after a few runs, I found k between 25 and 100 to usually be the best):

```python
from sklearn import neighbors
```

```python
print('Classifier 5: K-Nearest Neighbors')
print('')

clf = neighbors.KNeighborsClassifier()
param_grid = [{'n_neighbors': range(26,101,5)}] #, 'weights': ['uniform','distance']}]

bestkNNclf, train_acc, test_acc = gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test)

print('')

clfs[0,5] = train_acc
clfs[1,5] = test_acc
```

The best accuracy on the test set is obtained with 71 neighbors, which yields a 74.0% accuracy.

### f. Naive Bayes

The Naive Bayes classifier uses Bayes' rule in a simplified view where each feature is thought to be independent from the others. Even though this is an oversimplification, it has proven to be useful in many settings.
No hyperparameters need to be tuned.

```python
from sklearn import naive_bayes
```

```python
print('Classifier 6: Naive Bayes')
print('')

clf = naive_bayes.GaussianNB()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_train)
train_acc = accuracy_score(y_train, y_pred)

y_pred = clf.predict(X_test)
test_acc = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred)) 
print('Accuracy: %0.3f'%accuracy_score(y_test, y_pred))   
print('')

clfs[0,6] = train_acc
clfs[1,6] = test_acc
```

The accuracy is not as high as other models, with 67.7% on the test set.

### g. Decision tree and Random forest

The straightforward interpretability of decision trees make them very useful for analyzing the relevance of the different features. Random forests is the application of ensemble methods to decision trees. 

#### Decision tree

Let's start with a simple decision tree, with different values of max_depth:

```python
from sklearn import tree
```

```python
print('Classifier 7: Decision tree')
print('')

clf = tree.DecisionTreeClassifier(random_state=rng)
param_grid = [{'max_depth': [10,100,1000]}] # 'max_features': [1,2,7], 

bestDTclf, train_acc, test_acc = gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test)

print('')

clfs[0,7] = train_acc
clfs[1,7] = test_acc
```

#### Random forest

A random forest is basically an ensemble of decision trees, with average or majority voting. The higher number of decision trees are used, the lower the variance of the model is.

Increasing the number of decision trees to include (parameter n_estimators) will typically yield better performing models, but will take longer to train and evaluate, and will be heavier to store. For the sake of testing, let's try with 3 different values of n_estimators:

```python
from sklearn import ensemble
```

```python
print('Classifier 8: Random Forest')
print('')

clf = ensemble.RandomForestClassifier(random_state=rng, oob_score=True)
param_grid = [{'n_estimators': [10, 100, 300]}]

bestRFclf, train_acc, test_acc = gridSearch_thenTest(clf, param_grid, cv, X_train, y_train, X_test, y_test)

print('')

clfs[0,8] = train_acc
clfs[1,8] = test_acc
```

The RF with 300 estimators gives the best test accuracy, with 74.4%. Let's explore the model a little bit.

First, let's look at the importance of features averaged over all the trees:

```python
RFcoef = pd.DataFrame(np.reshape(bestRFclf.feature_importances_,(1,7)), columns=feats.columns[:-1])
RFcoef.T.plot(kind='barh')
plt.ylabel('RF feature importance value')
```

Although EOG_std and EEG_fractal_exponent are still the highest ranked features, the higher relative importance of band power features shows the random forest was able to use them and gain better accuracy.

## 3. Model selection

Now that we have tried multiple classifiers, let's compare them directly:

```python
clfsDF = pd.DataFrame(clfs, 
                      columns=['Random classifier','LDA','Logistic regression','LinearSVM','rbfSVM','kNN','Naive Bayes','Decision tree','Random forest'],
                      index=['train_acc','test_acc'])
clfsDF.head()
clfsDF.T.plot(kind='bar')
```

We see that the rbf-SVM is the best classifier for this task, but that the difference is very small between the best performing classifiers (logistic regression, kNN, RF, SVM). This overall similar performance of the classifiers shows that this classification problem is really a feature extraction one: to get significant increases in accuracy, better features must be used. To this end, additional features such as the other 21 included in the original study might be useful; also, learning features directly from the data by unsupervised pre-training might unveil new relevant features. Using more data, obviously, can also improve the results.

Let's see if these tweaks can increase our best accuracy:

### a. Dataset augmentation

As if often the case in EEG studies, ratios of band powers can be used to better inform us about ongoing mental activity. Let's compute additionnal features based on ratios with EEG_beta:

```python
augFeats = feats

augFeats['EEG_alpha/beta'] = feats.EEG_alpha/feats.EEG_beta
augFeats['EEG_theta/beta'] = feats.EEG_theta/feats.EEG_beta
augFeats['EEG_gamma/beta'] = feats.EEG_gamma/feats.EEG_beta
augFeats['EEG_gamma/delta'] = feats.EEG_gamma/feats.EEG_delta
```

Now, let's train our best model again with all these features, on the whole training set:

```python
# Separate data in train and test sets
X_train_aug, X_test_aug, y_train_aug, y_test_aug = cross_validation.train_test_split(                                    
                                                    augFeats.drop('label',axis=1), augFeats.label, test_size=0.15, random_state=rng)

rbfSVM_clfAug = bestKernelSVMclf
rbfSVM_clfAug.fit(X_train_aug, y_train_aug)

y_pred = rbfSVM_clfAug.predict(X_train_aug)
train_acc = accuracy_score(y_train_aug, y_pred)

y_pred = rbfSVM_clfAug.predict(X_test_aug)
test_acc = accuracy_score(y_test_aug, y_pred)
print(classification_report(y_test_aug, y_pred)) 
print('Validation accuracy: %0.3f'%train_acc)   
print('Test accuracy: %0.3f'%test_acc)   
print('')
```

This did not yield better results than using the original 7 features. Other transformations of the features (power, log, etc.) could be applied and might yield interesting results.

### b. Other strategies to improve the results

The obvious way of improving the results would be to use a larger subset of or the complete preprocessed data available (25 subjects, 28 features).

Similarly, this additionnal data could be used to perform unsupervised pre-training with stacked RBMs (like in the original paper) or with autoencoders trained on top of the pre-computed features. The currently unused classes (other stages of sleep) could thus be leveraged to improve the classification performance.

Finally, new EEG/EOG/EMG features could be extracted in addition to the ones already computed. This could also be done as showcased in the original paper by *learning* features with a similar unsupervised pre-training approach, but carried on the raw data. Different deep neural networks architectures such as convolutional networks could also be used on the temporal data (such as on the spectrogram of each channel).

## Conclusion

Classification of two consciousness states ('drowsy' and 'awake') was attempted on a subset of the UCD sleep database. Seven features extracted on 42,570 one-second windows were classified using 7 different classifiers:
1. LDA
2. Logistic regression
3. Linear SVM
4. Kernel SVM
5. k-NN
6. Naive Bayes
7. Decision tree
8. Random forest

The best performing classifier was the rbf-SVM with 75.6% accuracy on the test set. However, differences between classifiers were not huge. A nested cross-validation would allow to make sure this was the optimal classifier by giving a better estimate of the test accuracy (instead of a point estimate).

Using more/better features, such as hand-engineered or learned-from-data features, is probably the best way of increasing the accuracy of the proposed models.
